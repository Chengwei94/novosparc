```python
# imports
%matplotlib inline
import novosparc

import os
import numpy as np
import pandas as pd
import scanpy as sc
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist, squareform, pdist
from scipy.stats import ks_2samp
from scipy.stats import pearsonr
import anndata
import ot.bregman as ot
import io 
from contextlib import redirect_stdout


import random
random.seed(0)
```

```python
data_path = '/home/chengwei/Chengwei/10X_Data/Spatial/Parent_Visium_Human_BreastCancer/'
data = sc.read_visium(data_path)
data.var_names_make_unique()
```

```python
ad_sc = anndata.AnnData(data.X)
sc.pp.normalize_per_cell(ad_sc)
sc.pp.log1p(ad_sc)
```

```python
locations = data.obsm['spatial']
locations = locations / locations.max(axis = 0)

num_locations = locations.shape[0]
locations_circle = novosparc.gm.construct_circle(num_locations=num_locations)
```

```python
tit_size = 15
dpi = 100
fig, ax = plt.subplots(1,2, figsize=(12,3), dpi=dpi)

ax[0].scatter(locations[:, 0], locations[:, 1], s=3)
ax[1].scatter(locations_circle[:, 0], locations_circle[:, 1], s=3)

ax[0].set_title('Target space available apriori', size=tit_size)
ax[1].set_title('No prior knowledge of target space', size=tit_size)

plt.tight_layout()
plt.show()
```

```python
#dge_rep = None # a representation of cells gene expression
sc.pp.highly_variable_genes(ad_sc)
is_var_gene = ad_sc.var['highly_variable']
var_genes = list(is_var_gene.index[is_var_gene])
# alternative 1: variable expressed genes representation
dge_rep = ad_sc.to_df()[var_genes]
```

```python
tissue = novosparc.cm.Tissue(dataset=ad_sc, locations=locations_circle)
num_neighbors_s = num_neighbors_t = 5
# tissue.setup_reconstruction(num_neighbors_s=num_neighbors_s, num_neighbors_t=num_neighbors_t)
# alternative: since we don't have the atlas assumption, we can also setup only the smooth costs.
tissue.setup_smooth_costs(dge_rep=dge_rep, num_neighbors_s=num_neighbors_s, num_neighbors_t=num_neighbors_t)
```

```python
alpha_linear = 0 
random_ini = False
verbose = True
epsilon = 5e-3
loss_fun = 'kl_loss'
```

```python
tissue.reconstruct(alpha_linear=alpha_linear, epsilon=epsilon, verbose = True, loss_fun = loss_fun, iteration=200)
```

```python
gene_names = data.var.index.tolist()

sdge = tissue.sdge
dataset_reconst = sc.AnnData(pd.DataFrame(sdge.T, columns=gene_names))
dataset_reconst.obsm['spatial'] = locations
```

```python
# reconstructed expression of individual genes
pl_genes = ['EZH2', 'CD4', 'BRCA2', 'BRCA1']
novosparc.pl.embedding(dataset_reconst, pl_genes)
```
