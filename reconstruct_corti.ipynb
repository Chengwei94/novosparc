```python
# imports
from openpyxl import load_workbook
import novosparc
import scanpy as sc
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial.distance import pdist, squareform
from mpl_toolkits.axes_grid1 import make_axes_locatable

import warnings
warnings.filterwarnings("ignore")

import os 
outdir = 'output_corti/'
```

# Reconstruct Corti

De novo reconstruction of single-cell qRT-PCR measurements of the mammalian organ of corti, the receptor organ for hearing from [Waldhaus et al.](https://www.ncbi.nlm.nih.gov/pmc/articles/PMC4465070/). 

Specifically, we show:
1. recapitulating outer hair cell (OHC) subpopulation locations along the base-to-apex axis
2. OHC recovered expression
3. recapitulating of locations along the base-to-apex axis of other subpopulations

```python
# read data
fname = 'novosparc/datasets/corti/corti_expression.xlsx'

df_dge = pd.read_excel(fname, sheet_name='expression')
df_loc = pd.read_excel(fname, sheet_name='location')
df_type = pd.read_excel(fname, sheet_name='subpopulation')

df_dge.index = df_dge.index.astype('str')
df_loc.index = df_loc.index.astype('str')
df_type.index = df_type.index.astype('str')

df_dge_full = df_dge.copy()
df_loc_full = df_loc.copy()
df_type_full = df_type.copy()
```

## Recovering Outer Hair Cells(OHCs) Expression

We show first for a single subpopulation, OHC, that we manage to map cells to their anatomical origin and that we reconstruct the expression over the base and apex. 

```python
# subset na location and OHC subpopulation
genenames = list(df_dge.columns)
col_anat = df_loc.columns[0]

loc_not_na = ~df_loc[col_anat].isna()

col_subp = df_type.columns[0]
type_OHC = df_type[col_subp] == 'OHC'

idx = loc_not_na & type_OHC
df_dge = df_dge[idx]
df_loc = df_loc[idx]
df_type = df_type[idx]

_,ngenes = df_dge.shape
df_genes = pd.DataFrame(np.arange(ngenes), index=genenames, columns=['idx'])
```

```python
# preprocess expression
dataset = sc.AnnData(df_dge)
sc.pp.normalize_per_cell(dataset)
sc.pp.log1p(dataset)
```

```python
# construct target grid
ngrid = 10
grid_locs = np.vstack((range(ngrid),np.ones(ngrid))).T
```

```python
# construct tissue
tissue = novosparc.cm.Tissue(dataset=dataset, locations=grid_locs)
```

```python
# cost matrices
# tissue.setup_smooth_costs() # may want to remove graph construction
sp_expression = squareform(pdist( df_dge.values ,'correlation'))
cost_expression = sp_expression / sp_expression.max()
cost_expression -= np.mean(cost_expression)

num_locations = len(grid_locs)
sp_locations = np.zeros((num_locations,num_locations))
for i in range(num_locations):
    for j in range(num_locations):
        sp_locations[i,j] = np.abs(i-j)
    
cost_locations = sp_locations / sp_locations.max()
cost_locations -= np.mean(cost_locations)

tissue.costs['expression'] = cost_expression
tissue.costs['locations'] = cost_locations
```

```python
# reconstruction
alpha_linear = 0
tissue.reconstruct(alpha_linear=alpha_linear)
```

```python
# fetch results
gw = tissue.gw
sdge = tissue.sdge

df_gw = pd.DataFrame(gw, index=df_dge.index) # cells x grid
df_sdge = pd.DataFrame(sdge, index=genenames, columns=df_gw.columns) # grid x genes
```

### Recovering OHCs Anatomical Origin

Looking at the mapping of cells to the grid we constructed, we see a distinction between cells originating from the base and from the apex in terms of the layers they are mapped onto in the virtual grid.

```python
%matplotlib inline
# plot the grid layers apex and base cells are mapped onto

df_gw = pd.DataFrame(gw, index=df_loc.index) # cell to grid
col_grid = df_gw.columns
df_gw[col_anat] = df_loc[col_anat]
df_zone_to_grid = df_gw.groupby(col_anat)[col_grid].mean() # zone x grid layer
df_gw.drop(columns=col_anat, inplace=True)
df_zone_to_grid_norm = df_zone_to_grid.T.divide(df_zone_to_grid.max(1)).T

fonts = 20
fonts_ticks=15
plt.figure(figsize=(5,4))

plt.plot(df_zone_to_grid_norm.loc[['Base','Apex']].T)
plt.ylabel('Cellular membership', fontsize=fonts)
plt.xlabel('Embedded layer', fontsize=fonts)
plt.legend(('base','apex'), fontsize=fonts)
plt.yticks(fontsize=fonts_ticks)
plt.xticks(fontsize=fonts_ticks)
plt.show() 

fname = os.path.join(outdir, 'pred_OHC_to_grid.png')
plt.savefig(fname)
```

### Recovering Expression of Base and Apex

Here we see we manage to reconstruct the spatial expression profile of base and apex marker genes

```python
base_genes = ['Otof', 'Ptprq', 'Clrn1','Calb1', 'Calb2', 'Ocm']
apex_genes = ['Actb', 'Myo15', 'Pcdh15','Isl1','Dach1','Eya2','Pax2']

df_sdge_norm = df_sdge.T.divide(df_sdge.max(1)).T

fonts_leg = 13
plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.plot(df_sdge_norm.loc[base_genes].T)
plt.xlabel('Embedded layer', fontsize=fonts)
plt.ylabel('Expression', fontsize=fonts)
plt.title('Genes zonated towards base', fontsize=fonts)
plt.legend((base_genes), fontsize=fonts_leg)
plt.ylim(0,1)
plt.xticks(fontsize=fonts_ticks)
plt.yticks(fontsize=fonts_ticks)

plt.subplot(1,2,2)
plt.plot(df_sdge_norm.loc[apex_genes].T)
plt.xlabel('Embedded layer', fontsize=fonts)
plt.ylabel('Expression', fontsize=fonts)
plt.legend((apex_genes), fontsize=fonts_leg)
plt.title('Genes zonated towards apex', fontsize=fonts)
plt.ylim(0,1)
plt.yticks(fontsize=fonts_ticks)
plt.xticks(fontsize=fonts_ticks)

plt.tight_layout()
plt.show()

fname = os.path.join(outdir, 'pred_OHC_exp.png')
plt.savefig(fname)
```

## Mappings of All Subpopulations

Using few genes that differ in their expression within base and apex regions are sufficient for recovering locations of other subpopulations.

```python
# using ~differentially expressed genes
df_dge_full[col_anat] = df_loc_full[col_anat]
dge_full_mean = df_dge_full.groupby(col_anat)[genenames].mean()
df_dge_full.drop(columns=col_anat, inplace=True)

eps = 10**-6
ngenes_each = 30
apex_zonated = dge_full_mean.loc['Apex'].divide(dge_full_mean.loc['Base'] + eps).sort_values().index[-ngenes_each:]
base_zonated = dge_full_mean.loc['Base'].divide(dge_full_mean.loc['Apex'] + eps).sort_values().index[-ngenes_each:]
zonated_genes = list(set(apex_zonated).union(set(base_zonated)))

types = df_type_full[col_subp].unique()
ntypes = len(types)

ngrid = 5
grid_locs = np.vstack((range(ngrid),np.ones(ngrid))).T
anats = ['Base', 'Apex']
nanats = len(anats)
zone_to_grid_list = []


for i_ct, cell_type in enumerate(types):
#     print(cell_type)
    is_type = df_type_full[col_subp] == cell_type

    idx = loc_not_na & is_type
    df_dge = df_dge_full[idx]
    df_loc = df_loc_full[idx]

    dataset = sc.AnnData(df_dge)
    sc.pp.normalize_per_cell(dataset)
    sc.pp.log1p(dataset)
    
    tissue = novosparc.cm.Tissue(dataset=dataset, locations=grid_locs)
    
    sp_expression = squareform(pdist(df_dge[zonated_genes].values ,'correlation'))
    cost_expression = sp_expression / sp_expression.max()
    cost_expression -= np.mean(cost_expression)

    num_locations = len(grid_locs)
    sp_locations = np.zeros((num_locations,num_locations))
    for i in range(num_locations):
        for j in range(num_locations):
            sp_locations[i,j] = np.abs(i-j)

    cost_locations = sp_locations / sp_locations.max()
    cost_locations -= np.mean(cost_locations)

    tissue.costs['expression'] = cost_expression
    tissue.costs['locations'] = cost_locations
    
    alpha_linear = 0
    tissue.reconstruct(alpha_linear=alpha_linear, verbose=False)
    
    df_gw = pd.DataFrame(tissue.gw, index=df_loc.index) # cell to grid
    col_grid = df_gw.columns
    df_gw[col_anat] = df_loc[col_anat]
    df = df_gw.groupby(col_anat)[col_grid].mean()
    df['type'] = cell_type
    df['anat'] = df.index
    zone_to_grid_list.append(df.copy())
    df_gw.drop(columns=col_anat, inplace=True)
    
df_zone_to_grid = pd.concat(zone_to_grid_list).reset_index()
```
